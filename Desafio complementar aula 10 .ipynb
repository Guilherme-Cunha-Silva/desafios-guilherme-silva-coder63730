# Desafio complementar

1- Usando o arquivo "best-selling game consoles.xlsx" conte quantos consoles tem na base por companhias.
2- Ordene o resultado anterior pela quantidade de consoles por companhia de forma decrescente. 
3- Sobre o dataframe anterior, adicione a coluna "Market Value" presente na tabela do arquivo "company.txt".
4- Qual o valor m√©dio de Market Value por console?

```python
import pandas as pd
```

```python
df = pd.read_excel("best-selling game consoles.xlsx", sheet_name="consoles")
df.head()
```

```python
df1 = df.groupby('Company')['Console Name'].count()
df1
```

```python
df1 = df1.sort_values(ascending=False).reset_index(name='Contagem_Consoles')
df1
```

```python
df_company= pd.read_csv('company.txt', delimiter=',')
df_company.head()
```

```python
df_combinado = df1.merge(df_company, on='Company', how='left')
df_combinado.head()
```

```python
# Valor m√©dio de Market Value por console
media_console = df_combinado['Market Value'].sum() / df_combinado['Contagem_Consoles'].sum()

print(f"A m√©dia por console √©: {media_console:.2f}")
```
