# Desafio Nintendo

Abra o arquivo "best-selling game consoles.xlsx" em um dataframe.
Substitua a palavra "NES" por "Nitendinho" no nome dos consoles e deixe todos os nomes em maiúsculos.
Filtre o nome dos consoles com data de release depois de 2010.
De um describe e info da base, substitua os missing values pela string "missing".
Filtre os consoles que foram descontinuados a menos de 2 anos da data de release.

```python
import pandas as pd
```

```python
#Lendo o .xlsx
df = pd.ExcelFile('best-selling game consoles.xlsx')
```

```python
#Quais são as abas?
df.sheet_names
```

```python
#Criando o df da aba que quero
df_consoles = pd.read_excel('best-selling game consoles.xlsx', sheet_name='consoles')
```

```python
# Quais são as colunas?
df_consoles.columns
```

```python
#Olhando o tipo de informação de cada coluna 
df_consoles.dtypes
```

```python
# Printando as 5 linhas para olhar a tabela
print(df_consoles.head(5))
```

```python
# Substituindo a palavra "NES" por "Nitendinho" no nome dos consoles e deixando todos os nomes em maiúsculos.
df_consoles['Console Name'] = df_consoles['Console Name'].str.replace('NES','Nitendinho')
df_consoles['Console Name'] = df_consoles['Console Name'].str.upper()
```

```python
#Filtre o nome dos consoles com data de release depois de 2010.
df_consoles_filtrado = df_consoles[df_consoles['Released Year']>2010]
```

```python
df_consoles_filtrado
```

```python
#De um describe e info da base, substitua os missing values pela string "missing".
df_consoles_filtrado.describe(include='all')
```

```python
print(df_consoles_filtrado.info())
```

```python
# PROFF, não consegui aquir substituir pelo str.replace, pois ele substitui strings especificas né?.
    # Dessa forma, pesquisei na net e encontrei a função fillna
df_consoles_filtrado = df_consoles_filtrado.fillna("missing")
```
